# Korelasyon Nedir ?

- Korelasyon (bağıntı), iki değişkenin birbiriyle nasıl değiştiğini ölçen bir yöntemdir.
Örneğin, boy uzunluğu ile vücut ağırlığı arasındaki bağıntı, yaş ile boy arasındaki bağıntı, çalışma süreleri ve sınav başarıları arasındaki bağıntı veya hava sıcaklığı ile enerji tüketimi arasındaki bağıntı, eğer veriler sayısal ise korelasyon katsayısı ile ifade edilebilir.

- Korelasyon katsayısı, matematiksel olarak -1 ile +1 arasında değişen bir değer alır.

- Temel Özellikler:
Korelasyonun şiddeti (0-1), iki değişken arasındaki ilişkinin kuvvetini gösterir. İşaret (+ veya -) ise değişkenlerin aynı doğrultuda (+) ya da zıt yönlerde (-) hareket ettiğini ifade eder.
Örneğin, hava sıcaklığı ile doğalgaz kullanımı arasındaki bağıntı NEGATİF bir ilişki gösterebilir.
Çalışma süresi ile sınav başarısı arasında ise POZİTİF bir ilişki bulunabilir.

- Özel Durumlar:
Eğer iki değişken arasında bir bağ yoksa, korelasyon katsayısı sıfır veya sıfıra çok yakın bir değer alır.
İki değişken arasında tam bir bağ varsa, korelasyon katsayısı maksimum (+1) veya minimum (-1) değerine ulaşır ve bu durum kusursuz ilişkiyi ifade eder.

# Korelasyon Katsayısı Yorumu
- İki değişken arasında hesaplanan korelasyon (r)
değeri:

    • r<0.20 ve sıfıra yakın değerler ilişkinin olmadığı ya
    da çok zayıf ilişkiyi işaret eder.

    • 0.20-0.39 arasında ise zayıf ilişki

    • 0.40-0.59 arasında ise orta düzeyde ilişki

    • 0.60-0.79 arasında ise yüksek düzeyde ilişki

    • 0.80-1.0 ise çok yüksek ilişki olduğu yorumu
    yapılır.

# Korelasyon İlişkilerinin Yönü
Korelasyon, iki değişken arasındaki ilişkinin yönünü (pozitif veya negatif) belirler. Bu yön, değişkenlerin birbirine nasıl tepki verdiğini açıklar:

1. Pozitif Korelasyon (+)
Tanım: İki değişken aynı yönde hareket eder.
Bir değişken arttığında, diğeri de artar.
Bir değişken azaldığında, diğeri de azalır.
Örnekler:
Çalışma süresi arttıkça sınav puanlarının artması.
Boy uzunluğu arttıkça kilo artışı gözlenmesi.
2. Negatif Korelasyon (-)
Tanım: İki değişken zıt yönde hareket eder.
Bir değişken arttığında, diğeri azalır.
Bir değişken azaldığında, diğeri artar.
Örnekler:
Hava sıcaklığı arttıkça doğalgaz tüketiminin azalması.
Bir ürünün fiyatı arttıkça talebin azalması.
3. Korelasyon Yönsüz (0 veya Sıfıra Yakın)
Tanım: İki değişken arasında anlamlı bir ilişki yoktur.
Değişkenlerden birinin değişimi, diğerini etkilemez.
Örnekler:
Bir kişinin ayakkabı numarası ile sınav puanı arasında ilişki olmaması.

# 1) Pearson Korelasyon Katsayısı

- Pearson korelasyonu, iki değişken arasındaki doğrusal ilişkiyi ölçer. Bu nedenle her iki değişkenin de kesikli (ör. sayılar) veya sürekli (ör. ölçümler) sayısal veri olması gerekir.
- Pearson korelasyonu yalnızca değişkenler arasındaki ilişki doğrusal olduğunda anlamlıdır.
- Her iki değişkenin de yaklaşık olarak normal dağılım göstermesi beklenir.
Veriler normal dağılmıyorsa, Spearman sıralama korelasyonu gibi alternatif yöntemler tercih edilebilir.
- Özetle iki veri arasında doğrusal bir ilişki olup olmadığına bakar.
- H0: İki örnek bağımsızdır.(**ρ=0**)
- H1: İki örnek bağımlıdır.(**ρ≠0**)

```python
import pandas as pd
import pingouin as pg
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import normaltest
import matplotlib.pyplot as plt
# GEREKLİ KÜTÜPHANELER YÜKLENDİ
```

```python
data=pd.read_csv('pearson_data.csv')

print(data)  
```

```python
## VERİMİZİ GÖRSELLEŞTİRELİM

plt.plot(data['Reklam_Harcaması'], data['Satıs_Geliri'])  
plt.show()
```

```python
## VERİMİZİN NORMAL DAĞILIP DAĞILMADIĞINI PİNGOUİN KÜTÜPHANESİ KULLANARAK KONTROL EDELİM

## PİNGOUİN KÜTÜPHANESİ VERİNİN NORMAL DAĞILIP DAĞILMADIĞINI KONTROL ETMEK İÇİN VARSAYILAN OLARAK Shapiro-Wilk TESTİNİ KULLANIR


normallik=pg.normality(data)   ## çıkan sonuçlara göre kullanacağımız veriler normal dağılıyor
print(normallik) 
```

```python
## SCİPY KÜTÜPHANESİ ÜZERİNDEN NORMALLİK TESTİ YAPALIM
## SCİPY KÜTÜPHANESİ VERİNİN NORMAL DAĞILIP DAĞILMADIĞINI KONTROL ETMEK İÇİN VARSAYILAN OLARAK  D'Agostino TESTİNİ KULLANIR
## NORMAL DAĞILIM VE NORMALLİK TESTLERİ HAKKINDA DAHA FAZLA BİLGİYE DİĞER NOTLARIM ARASINDAN ULAŞABİLİRSİNİZ

print(normaltest(data['Reklam_Harcaması'])) ## pvalue DEĞERİ 0.05'ten büyük veri normal dağılıyor
print(normaltest(data['Satıs_Geliri'])) ## pvalue DEĞERİ 0.05'ten büyük veri normal dağılıyor
```

```python
## DOGRUSALLIK TESTİ YAPALIM 
## ÖNCELİKLE BAĞIMLI DEĞİŞKEN VE BAĞIMSIZ DEĞİŞKENLERİ BELİRLEYELİM
## BAĞIMLI DEĞİŞKEN x=SATIŞ GELİRİ
## BAĞIMSIZ DEĞİŞKEN y=REKLAM HARCAMASI

## Aşağıdaki kod, Reklam Harcaması ve Satış Geliri arasındaki ilişkiyi görsel olarak incelememizi sağlamak için kullanılmıştır.
## Kodun çıktısında görülen regresyon doğrusu, iki değişken arasındaki doğrusal ilişkiyi gösterir.
## Bu görselleştirme, ilişkilerin doğrusal olup olmadığını anlamamıza yardımcı olur. Şu an için sadece ilişkilerin doğrusal olduğunu gözlemliyoruz."
## REGRESYON ANALİZİ HAKKINDA DAHA FAZLA BİLGİYE DİĞER NOTLARIM ARASINDAN ULAŞABİLİRİSİNİZ

sns.lmplot(x= 'Reklam_Harcaması' , y='Satıs_Geliri',data=data)
plt.show()


## ÇIKTIDA GÖRÜLDÜĞÜ ÜZERE İKİ VERİ ARASINDA DOĞRUSAL İLİŞKİ VAR
```

```python
## VERİLER NORMAL DAĞILIYOR VE ARALARINDA DOĞRUSAL İLİŞKİ VAR ŞİMDİ KORELASYONU HESAPLAYALIM

## HİPOTEZLER 
#  H0: REKLAM HARCAMASI VE SATIŞ GELİRİ ARASINDA ANLAMLI İLİŞKİ YOKTUR 
#  H1: REKLAM HARCAMASI VE SATIŞ GELİRİ ARASINDA ANLAMLI İLİŞKİ VARDIR


korelasyon=pg.corr(x=data['Reklam_Harcaması'],y=data['Satıs_Geliri'],method='pearson')

print(korelasyon) 

### SONUCU YORUMLAYALIM 
# 1-) r yani korelasyon katsayısı pozitif ve 1'e çok yakın iki değişken arasında pozitif yönlü güçlü bir ilişki var.
# 2-) p-value değeri 0.05 ten küçük H0 hipotezi reddedildi.
```

# 2) Spearman Korelasyon Katsayısı

- Monotonik İlişki:

Spearman korelasyonu, monotonik ilişkiyi test eder. Bu, değişkenlerden birinin arttıkça diğerinin de artması (pozitif monoton) veya birinin arttıkça diğerinin azalması (negatif monoton) anlamına gelir.
Monotonik ilişki, doğrusal olmak zorunda değildir; sadece artış veya azalış düzeniyle bir ilişki olmalıdır.

- Sıralı (Ordinal) veya Sürekli (Interval/Ratio) Veriler:

Spearman korelasyonu, sıralı veriler (ordinal) veya sürekli veriler ile çalışabilir. Ancak verilerin sıralı (yani, ölçüm sırası) olması gerekir.
Örneğin, her iki değişkenin de sıralı veri (örneğin, sıralama veya derecelendirme) ya da sürekli (gerçek sayılar) olması gereklidir.

-  Doğrusallık varsayımı yoktur.
-  Veri Normal Dağılmıyorsa Kullanılır.
- H0: İki örnek bağımsızdır.(**ρ=0**)
- H1: İki örnek bağımlıdır.(**ρ≠0**)


```python
## VERİMİZİ YAZDIRALIM

data=pd.read_csv('spearman_data.csv')
print(data)
```

```python
## VERİMİZİ GÖRSELLEŞTİRELİM


plt.plot(data['Variable_X'], data['Variable_Y'])  
plt.xlabel("Variable_X")  
plt.ylabel("Variable_Y")  
plt.title("Çizgi Grafiği") 
plt.show()
```

```python
## NORMALLİK TESTİ YAPALIM


normallik=pg.normality(data)  ## VERİLERDEN BİRİ NORMAL DAĞILMIYOR
print(normallik)
```

```python
## DOĞRUSALLIK TESTİ YAPALIM 

sns.lmplot(x='Variable_X',y='Variable_Y',data=data) ## Çıktıda Görüldüğü Üzere Veriler Arasında Doğrusal İlişki Yok. Gölgeli kısım güven aralığını gösterir
plt.show()
```

```python
## VERİLERDEN BİRİ NORMAL DAĞILMIYOR VE ARALARINDA DOĞRUSAL İLİŞKİ YOK ŞİMDİ KORELASYONU HESAPLAYALIM

## HİPOTEZLER 
#  H0: X DEĞERLERİ İLE Y DEĞERLERİ  ARASINDA ANLAMLI İLİŞKİ YOKTUR 
#  H1: X DEĞERLERİ İLE Y DEĞERLERİ  ARASINDA ANLAMLI İLİŞKİ VARDIR




korelasyon=pg.corr(x=data['Variable_X'],y=data['Variable_Y'],method='spearman')
print(korelasyon)



### SONUCU YORUMLAYALIM 
# 1-) r yani korelasyon katsayısı pozitif ve 1'e çok yakın iki değişken arasında pozitif güçlü yönlü bir ilişki var.
# 2-) p-value değeri 0.05 ten küçük H0 hipotezi reddedildi.
```
